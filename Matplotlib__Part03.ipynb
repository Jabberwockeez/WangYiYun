# 【课程3.13】  表格样式创建

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
sty = df.style
print(sty)
sty
```

```python
def color_neg_red(val):
    if val < 0:
        color = 'red'
    else:
        color = 'black'
    return('color:%s' % color)
df.style.applymap(color_neg_red)
```

```python
def highlight_max(s):
    is_max = s == s.max()
    lst = []
    for v in is_max:
        if v:
            lst.append('background-color: yellow')
        else:
            lst.append('')
    return lst
df.style.apply(highlight_max, axis=0, subset=list('ABD'))
```

```python
df.style.apply(highlight_max, axis=0, subset=pd.IndexSlice[2:5, list('AB')])
```

# 【课程3.15】  表格样式调用

```python
df = pd.DataFrame(np.random.rand(5, 4), columns=list('ABCD'))
df['A'][2] = np.nan
df.style.highlight_null(null_color='red')
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list('ABCD'))
df.style.background_gradient(cmap='Greens', axis=1, low=0, high=1)
```

```python
df.style.bar(subset=['A', 'B'], color='#d65f5f', width=100)
```

```python
df = pd.DataFrame(np.random.rand(10,4),columns = list('ABCD'))
df['A'][[3, 2]] = np.nan
df.style.\
    bar(subset=['A', 'B'], color='#D65F5F', width=100).\
    highlight_null(null_color='green')
```
