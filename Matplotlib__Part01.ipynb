# 【课程3.2】  图表的基本元素

# 【课程3.4】  刻度、注解、图表输出

```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

```python
t = np.arange(0.0, 100.0, 1)
s = np.sin(0.1 * np.pi * t) * np.exp(-t * 0.01)
ax = plt.subplot(111)
plt.plot(t, s, '--.')
plt.grid(True, linestyle=':', color='gray', linewidth='0.3')
# plt.legend(loc=0)   # 无效

xmajorLocator = MultipleLocator(20)          # 将x主刻度标签设置为10的倍数
xmajorFormatter = FormatStrFormatter('%.0f') # 设置x轴标签文本的格式
xminorLocator = MultipleLocator(15)          # 将x次刻度标签设置为5的倍数

ax.xaxis.set_major_locator(xmajorLocator)    # 应用x轴主刻度
ax.xaxis.set_major_formatter(xmajorFormatter)# 应用x轴标签文本格式
ax.xaxis.set_minor_locator(xminorLocator)    # 应用x轴次刻度

ax.xaxis.grid(True, which='minor')

# ax.xaxis.set_major_locator(plt.NullLocator())  # 删除x轴主刻度
```

```python
df = pd.DataFrame(np.random.rand(10, 2))
fig = df.plot(style='--o', figsize=[8,6])
ax_1 = plt.subplot(111)
plt.legend(loc=0)
plt.grid(True, linestyle=':', color='purple', linewidth='0.3')
plt.text(5, 0.5, "That's IT", fontsize=8)

plt.yticks([0, 0.2, 0.4, 0.5, 0.6, 0.8, 1.0])
fig.set_yticklabels('%.1f' %i for i in [0, 0.2, 0.4, 0.5, 0.6, 0.8, 1.0])
print(fig)

ymajorLocator = MultipleLocator(0.5)
ymajorFormatter = FormatStrFormatter('%.2f')
yminorLocator = MultipleLocator(0.2)

ax_1.yaxis.set_major_locator(ymajorLocator)
ax_1.yaxis.set_major_formatter(ymajorFormatter)
ax_1.yaxis.set_minor_locator(yminorLocator)

ax_1.yaxis.grid(True, which='both')
```

```python
fig = plt.figure(figsize=(10,8), facecolor='gray')

ax_1 = fig.add_subplot(2,2,1)
plt.plot(np.random.rand(50).cumsum())
plt.plot(np.random.randn(50).cumsum())
# plt.legend(loc=0)

ax_2 = fig.add_subplot(2,2,2)
ax_2.hist(np.random.rand(50), alpha=0.6)

ax_4 = fig.add_subplot(2,2,4)
df = pd.DataFrame(np.random.rand(10,4), columns=list('ABCD'))
plt.plot(df, alpha=0.3, linestyle='--', marker='.')
# plt.legend(loc=0)
```
